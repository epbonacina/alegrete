### Visualizacao dos dados

### Execucao da regressao linear e calculo do EQM por epoca/iteracao

```python
theta_0s, theta_1s = alegrete.fit(
    quiz_data, theta_0=0, theta_1=0, 
    alpha=0.001, num_iterations=50000
)

# valores finais de theta0 e theta1
final_th0, final_th1 = theta_0s[-1], theta_1s[-1]

mse_list = [alegrete.compute_mse(th0, th1, quiz_data) for th0, th1 in zip(theta_0s, theta_1s)]
print(f"EQM final: {mse_list[-1]}")


plt.plot(mse_list)
plt.xlabel('Epoca/iteracao')
plt.ylabel('EQM')
plt.show()
```

# Prova de otimalidade
Podemos resolver o problema com algebra linear.

```python
lsm = [[quiz_data.shape[0], quiz_data[:,0].sum()],
       [quiz_data[:,0].sum(), (quiz_data[:,0]**2).sum()]]
lsm = np.asarray(lsm, dtype=np.float64)
b = [quiz_data[:, 1].sum(), (quiz_data[:, 0] * quiz_data[:, 1]).sum()]
b = np.asarray(b, dtype=np.float64)
solution = np.linalg.solve(lsm, b)
print(solution)
```

```python
alegrete.compute_mse(solution[0], solution[1], quiz_data)
```
